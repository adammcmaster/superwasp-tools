```python
import pandas
import seaborn
from matplotlib import pyplot

import swasputils
```

```python
classifications = swasputils.AggregatedClassifications()
```

```python
classifications.add_classification_labels()
```

```python
seaborn.displot(
    data=classifications.df,
    x='Period',
    hue='Classification Label',
    kind='kde',
    log_scale=True,
)
pyplot.title('Distribution of Periods')
```

```python
for class_key, class_label in swasputils.AggregatedClassifications.CLASSIFICATION_LABELS.items():
    class_subset = classifications.get_class(class_key)
    seaborn.displot(
        data=class_subset.df,
        x='Period',
        kind='kde',
        log_scale=True,
    )
    pyplot.title('Distribution of Periods ({})'.format(class_label))
```

```python
for class_key, class_label in swasputils.AggregatedClassifications.CLASSIFICATION_LABELS.items():
    class_subset = classifications.get_class(class_key)
    seaborn.displot(
        data=class_subset.df,
        x='Period',
        kind='kde',
        log_scale=True,
        hue='Period Uncertainty'
    )
    pyplot.title('Distribution of Periods ({})'.format(class_label))
```
